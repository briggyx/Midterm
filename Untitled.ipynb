```python
import numpy as np
import pandas as pd
from tensorflow import keras

data = pd.read_csv('Final_Training_Data_with_Weather_Trimmed.csv')
data = data[['julian', 'crs_arr_time', 'crs_dep_time', 'crs_elapsed_time', 'distance', 'dep_temp', 'dep_rhum', 'dep_prcp',
             'dep_wdir', 'dep_wspd', 'dep_pres', 'arr_temp', 'arr_rhum', 'arr_prcp', 'arr_wdir', 'arr_wspd', 'arr_pres',
             'cat_ontime','cat_delayed_15','cat_delayed_60','origin','dest','dep_datetime_rounded','arr_datetime_rounded']]
data = data.sample(1000000)
```

```python
origin_onehot = pd.get_dummies(data.origin)
dest_onehot = pd.get_dummies(data.dest)

origin_onehot.columns = [ f'O_{x}' for x in origin_onehot.columns ]
dest_onehot.columns = [ f'D_{x}' for x in dest_onehot.columns ]

labels = data[['cat_ontime','cat_delayed_15','cat_delayed_60']]

# data = data[['julian', 'crs_arr_time', 'crs_dep_time', 'crs_elapsed_time', 'distance']]

data = data[['julian', 'crs_arr_time', 'crs_dep_time', 'crs_elapsed_time', 'distance', 'dep_temp', 'dep_rhum', 'dep_prcp',
             'dep_wdir', 'dep_wspd', 'dep_pres', 'arr_temp', 'arr_rhum', 'arr_prcp', 'arr_wdir', 'arr_wspd', 'arr_pres']]

for c in data.columns:
    data[c] = data[c] - np.min(data[c])
    data[c] = data[c] / np.max(data[c])

data = pd.merge(data,origin_onehot,left_index=True,right_index=True)
data = pd.merge(data,dest_onehot,left_index=True,right_index=True)

data = data.to_numpy(dtype='float32')
labels = labels.to_numpy(dtype='float32')
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(128, input_shape=(655,)))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(64))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(32))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(3,activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(data[:100000,:], labels[:100000,:], validation_data=(data[100000:150000,:], labels[100000:150000,:]), epochs=1000, batch_size=100000)
```
