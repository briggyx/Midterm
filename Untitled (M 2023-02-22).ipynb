```python
# Get a list of dataframes for each CSV file in the data subfolder

data = []

for x in os.listdir('data'):
    if '.csv' not in x:
        continue
    print(x)
    data.append(pd.read_csv(f'data/{x}',encoding='unicode_escape'))
```

```python
# Merge the list of dataframes into one, on columns
data = pd.concat(data)
```

```python
# Replace date string data with actual datetime data
dates = pd.to_datetime(data.fl_date)
years = dates.dt.year
julian_dates = dates.dt.day_of_year
```

```python
data['year'] = years
data['julian'] = julian_dates
```

```python
# Function to take an integer representing 24 hour time and convert it to a minute-of-the-day integer; front-end zero-padding needed

def hhmm_to_minutes(hhmm):
    hhmm = str(hhmm)
    hhmm = hhmm.split('.')[0]
    if len(hhmm) < 4:
        hhmm = '0' * (4-len(hhmm)) + hhmm
    hh = int(hhmm[:2]) * 60
    mm = int(hhmm[2:])
    hhmm = hh+mm
    return hhmm
```

```python
# Make new dataframe columns for minute-of-the-day for CRS departure and arrival times

data.crs_dep_time_minutes = data.crs_dep_time.apply(hhmm_to_minutes)
data.crs_arr_time_minutes = data.crs_arr_time.apply(hhmm_to_minutes)
```

```python
# Calculate the z-scores for arr_delay
z_scores = zscore(data_relevant['arr_delay'])

# Count the number of rows where |z-score| > 3
num_outliers = (abs(z_scores) > 3).sum()

# Calculate the percentage of rows that are outliers
pct_outliers = num_outliers / len(data_relevant) * 100

print(f"Percentage of rows where arr_delay has an absolute z-score greater than 3: {pct_outliers:.2f}%")
```
